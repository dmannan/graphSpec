### Create a multigraph csv

```python
with open("multiEdgelist.csv",'w') as new_file:
    
    csv_writer = csv.writer(new_file, delimiter = ',')
    
    csv_writer.writerow([1,2,3,6])
    csv_writer.writerow([2,3,8,7])
    csv_writer.writerow([3,4,5,8])
```

```python
data3 = pd.read_csv("multiEdgelist.csv", header = None)
data3
```

```python
g3 = nx.read_edgelist("multiEdgelist.csv", delimiter=',', data=(('eweight',float),('gweight', float)))
g3
```

```python
print(g3.edges)
g3.nodes
```

```python

g3.get_edge_data("3","4")
```

```python
array3 = nx.to_numpy_array(g3, weight='eweight') ##can make an array with one type of weight
array3
```

```python
array3b = nx.to_numpy_array(g3,weight='gweight')### can make an array with the other type of weight
array3b
```

### Trial of multiweight graph with None 

```python
with open("multiEdgelistNone.csv",'w') as new_file:
    
    csv_writer = csv.writer(new_file, delimiter = ',')
    
    csv_writer.writerow([1,2,3,6,9])
    csv_writer.writerow([2,3,8,7,NaN])
    csv_writer.writerow([3,4,5,NaN,8])
```

```python
data4 = pd.read_csv("multiEdgelistNone.csv", header = None)
data4
```

```python
g4 = nx.read_edgelist("multiEdgelistNone.csv", delimiter=',', data=(('eweight',float),('gweight', float),("fweight", float)))
g4.get_edge_data("3","4")
```

```python
G4 = nx.MultiGraph(g4)
array4 = nx.to_numpy_array(G4, weight = 'fweight')
array4
```
